### Soru 2:

Veri çerçevelerinin .groupby() metodunu kullanarak yolcuların ortalama yaşlarını herbir cinsiyet için ayrı olacak şekilde hesaplayın.

```python
soru2 = pd.read_csv("train.csv", index_col=0)    # Dosya yolunun, Windows'taki gibi backslash (\) ile yazılmaması gerekiyor.
soru2
```

```python
# pd.DataFrame( soru2.groupby("Sex").mean()["Age"])    ## bu kod, aşağıdakinden daha yavaş çalışıyor çünkü hepsinin ortalamasını alıp sonra Yaş'ı seçiyor.
pd.DataFrame( soru2.groupby("Sex")["Age"].mean())    ## bu kod ise Yaş'ı seçtikten sonra onun ortalamasını alıyor
```

### Soru 3:
Koşullu seçim yöntemlerini de kullanarak, 30 yaşından küçük tüm yolcuların kurtulma oranını (survival ratio) hesaplayın. Daha sonra, kurtulma oranını cinsiyete göre ayrı ayrı hesaplayın.

```python
all_30 = soru2[soru2["Age"]<30]
male_30 = soru2[(soru2["Age"]<30) & (soru2["Sex"]=="male")]    #-->  data frame'leri parantez () içine almayınca hata verdi.
female_30 = soru2[(soru2["Age"]<30) & (soru2["Sex"]=="female")]    # maalesef "&" yerine "and" işe yaramıyor. 

# all30_survival = pd.DataFrame(all_30.groupby("Sex").mean()["Survived"])   ## bu kod, aşağıdakinden daha yavaş çalışıyor çünkü hepsinin ortalamasını alıp sonra Kurtulanlar arasından seçiyor.
all30_survival = pd.DataFrame(all_30.groupby("Sex")["Survived"].mean())    ## bu kodda ise Kurtulanları seçip, sadece onların ortalamasını alıypr.
all30_survival.columns = ["Survival Ratio"]
all30_survival
```

```python
pd.DataFrame({"Survival Ratio":(male_30.groupby("Sex")["Survived"].mean())})    ## yukarıda yazdığım yavaşlık olayı bunda da düzeltildi.
```

```python
pd.DataFrame({"Survival Ratio":female_30.groupby("Sex")["Survived"].mean()})    # tek satıra düşürüldü.    ## yavaşlık düzeltildi.

# eski versiyon: 
# female30_survival = pd.DataFrame(female_30.groupby("Sex").mean()["Survived"])
# female30_survival.columns = ["Survival Ratio"]
# female30_survival    # Bu çıktıyı, 3 satır yerine tek satır kod ile yapmak mümkün mü?
```
